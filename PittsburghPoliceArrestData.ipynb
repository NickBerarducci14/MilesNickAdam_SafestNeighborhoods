**Introduction** - For our project on finding the best neighborhood in Pittsburgh, I focused on looking at safety using the Pittsburgh Police Arrest dataset. Since the dataset includes every recorded arrest along with the neighborhood where it happened, it gave me a good way to compare different areas. By counting how often each neighborhood showed up and turning that into a bar graph, I was able to see which places had higher or lower levels of police activity. This helped me get a better sense of how safety might differ across the city and which neighborhoods could be considered safer based on arrest frequency.

```python
import pandas as pd
import matplotlib.pyplot as plt

data_set = pd.read_csv("Pittsburgh_Police_Arrest_Data.csv")

#Looking at the column I want from the csv file
col = 'INCIDENTNEIGHBORHOOD'

#Getting the least amount of arrests in the neighborhoods to see which one is the safest
arrest_counts = data_set[col].value_counts().tail(10)

#Creating the Bar Graph
arrest_counts.plot.bar(color='blue')
plt.title("Safest Neighborhoods in Pittsburgh")
plt.xlabel("Arrest Location (Neighborhood)")
plt.ylabel("Frequency (Amount of Arrests)")
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Submetric conclusion** - Using the Pittsburgh Police Arrest dataset made it easy to see which neighborhoods show up the least when it comes to arrests. By counting how often each neighborhood appeared and graphing it, I was able to get a clearer picture of which area might be considered safer based on lower arrest numbers. Its definitely not a perfect measure of safety, but the dataset gives a solid starting point. As we can see, the least amount of arrests was Mt.
